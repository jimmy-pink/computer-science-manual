## 投票算法
Boyer-Moore Majority Vote Algorithm  

用于在数组中找到 **元素数量过半的那个元素** 的算法。


### Leet 169
image.png

```python
from typing import List
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        candidate = nums[0]
        count = 0
        for n in nums:
            if count == 0:
                candidate = n
            count += (1 if candidate == n else -1)
        return candidate

# 为候选人 2 计数相同+1，不同-1，减为0时，候选人指定为下一个, 也就是下标=4的那个1
arr = [2,2,1,1,1,2,2]
print("多数元素：", Solution().majorityElement(arr))
```
